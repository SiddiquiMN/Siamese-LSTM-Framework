```python
import os
import cv2
import numpy as np
import concurrent.futures

# Directory containing your preprocessed green images
images_directory = '/content/drive/MyDrive/Sowmik Dataset/New dataset/Residence_all_processed_data/pre_img_out'

# Directory to save binary masks
masks_directory = '/content/drive/MyDrive/Sowmik Dataset/New dataset/Residence_all_processed_data/bmask_residence'

# Load image filenames
image_filenames = [filename for filename in os.listdir(images_directory) if filename.endswith('.jpg')]

# Ensure masks directory exists
os.makedirs(masks_directory, exist_ok=True)

# Define a function to create binary masks
def create_binary_mask(filename):
    image_path = os.path.join(images_directory, filename)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    grayscale_lower = 0.2989 * 40 + 0.5870 * 40 + 0.1140 * 40
    grayscale_upper = 0.2989 * 80 + 0.5870 * 255 + 0.1140 * 255

    # Calculate the threshold value
    #threshold = (grayscale_lower + grayscale_upper) / 2
    threshold = 40  # Adjust this threshold as needed
    binary_mask = (image > threshold).astype(np.uint8) * 255

    mask_filename = filename.replace('.jpg', '_mask.jpg')
    mask_path = os.path.join(masks_directory, mask_filename)
    cv2.imwrite(mask_path, binary_mask)

batch_size = 100

# Use ThreadPoolExecutor for parallel processing
with concurrent.futures.ThreadPoolExecutor(max_workers=4) as executor:
    executor.map(create_binary_mask, image_filenames)
```
