```python
import os
import cv2
import numpy as np
```

```python
input_folder = "/content/drive/MyDrive/images"
```

```python

output_folder = "/content/drive/MyDrive/preprocessed_image"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through each image in the input folder
for image_filename in os.listdir(input_folder):
    if image_filename.endswith(".jpg"):  # Adjust file extension based on your image format
        image_path = os.path.join(input_folder, image_filename)

        # Read the image using OpenCV
        image = cv2.imread(image_path)


        hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)


        lower_green = np.array([40, 40, 40])
        upper_green = np.array([80, 255, 255])


        green_mask = cv2.inRange(hsv_image, lower_green, upper_green)


        green_highlighted_image = cv2.bitwise_and(image, image, mask=green_mask)


        output_path = os.path.join(output_folder, f"green_highlighted_{image_filename}")
        cv2.imwrite(output_path, green_highlighted_image)

print("Processing complete.")
```
